```python
from transformers import pipeline

# For a public model (e.g., sentiment analysis)
classifier = pipeline("sentiment-analysis")
result = classifier("Hugging Face models in Colab are great!")
print(result)
```

```python
!pip install gradio
```

```python
import gradio as gr



# Create the interface
# inputs='text' specifies a textbox input
# outputs='text' specifies a textbox output
demo = gr.Interface(fn=classifier, inputs="text", outputs="text")

# Launch the interface
# The launch() method starts a local server
# You can access the interface at the displayed URL
demo.launch()
```
